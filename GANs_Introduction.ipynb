### Representation learning

Learn to form and group feature-characteristics of a highly dimensional space (images)in a space of less dimensions (latent space). It learns a mapping function that takes points from the latent space to points in the original domain. A point in latent space is a representation of a point in high dimensional space.

### Add more smile to image in simple words:

take all the images with label smile in the dataset, map them to latent space and calculate the average of their position in latent space. Then substract this mean from the position of images which do not smile. It will give us the vector than points from the non-smile to the smile.

number of bts required t encode and transmit an event.

2. Higher probability, events have less information, they are less suprising.

## Entropy:
number of bits required to represent a randomly selected event from a probability distribution.

* Skewed distribution has low entropy - it it unsuprising
* uniform distribution has larer entropy, it is suprising.

H(X) = -sum{P(xi)log(P(xi))}

## Cross-entropy
number of bits required to represent an average event from one ditribtuion compared to another.

cross-entropy is the number of extra bits needed to represent an event using Q instead of P.

P = target distribition
Q = approimation of P

H(P,Q) = -sum{P(xi)log(Q(xi))}
